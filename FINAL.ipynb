## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create the bins in which Data will be held
bins = [0, 9.9, 13.9, 18.9, 23.9, 29.9, 33.9, 38.9,200]
# Create the names for the five bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39", "40+"]
#take the purchase data and add a column called age group that uses the bins/groups above
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_names, include_lowest=True)

# Create a GroupBy object based upon "Age Group"
age_group = purchase_data.groupby("Age Group")
# Find how many rows fall into each bin
age_group_df = pd.DataFrame(age_group["Purchase ID"].count())

# Using .rename(columns={}) in order to rename columns
age_group_df = age_group_df.rename(columns={"Purchase ID":"Total Count"})


age_group_df["Percentage of Players"] = age_group_df.loc[:, ["Total Count"]]/total_players
age_group_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# bin the purchase_data
purchase_data["Ranges of Age"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
# purchase_data.head()

age_analysis_total = purchase_data.groupby(["Ranges of Age"]).count()["Price"].rename("PURCHASE COUNT")

age_average_price = purchase_data.groupby(["Ranges of Age"]).mean()["Price"].rename("AVERAGE PURCHASE PRICE")

age_total_purchase = purchase_data.groupby(["Ranges of Age"]).sum()["Price"].rename("TOTAL PURCHASE VALUE")

age_totalpurchase_perperson = age_total_purchase / age_analysis_total


age_purchasing_analysis = pd.DataFrame({"PURCHASE COUNT": age_analysis_total,
                              "AVERAGE PURCHASE PRICE": age_average_price,
                              "TOTAL PURCHASE VALUE": age_total_purchase,
                              "AVG TOTAL PURCHASE PER PERSON": age_totalpurchase_perperson})

age_purchasing_analysis["AVERAGE PURCHASE PRICE"] = age_purchasing_analysis["AVERAGE PURCHASE PRICE"].map("${:.2f}".format)
age_purchasing_analysis["TOTAL PURCHASE VALUE"] = age_purchasing_analysis["TOTAL PURCHASE VALUE"].map("${:.2f}".format)
age_purchasing_analysis["AVG TOTAL PURCHASE PER PERSON"] = age_purchasing_analysis["AVG TOTAL PURCHASE PER PERSON"].map("${:.2f}".format)
                                                                                                                    

age_purchasing_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#This is grouping the main data by SN, and counting how many times they appear to show how many items they bought
sn_anal_tot = purchase_data.groupby(["SN"]).count()["Price"].rename("Purch Count")
#This is grouping the main data by SN and adding up the prices they spent
sn_anal_purchasetot = purchase_data.groupby(["SN"]).sum()["Price"]
#This is grouping the main data by SN and averageing the purchases
sn_anal_purchaseavg = purchase_data.groupby(["SN"]).mean()["Price"]
sn_purch_anal = pd.DataFrame({
     "Purchase Count": sn_anal_tot,
    "Average Purchase Price":sn_anal_purchaseavg,
    "Total Purchase Value":sn_anal_purchasetot
    })
sn_purch_anal=sn_purch_anal.sort_values(by='Total Purchase Value', ascending=False)
sn_purch_anal.head(10)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_df = purchase_data.loc[:,["Item Name","Item ID", "Price"]]

groupedby_popular = popular_df.groupby(["Item ID","Item Name"])
groupedby_popular.head()
#do the counts
pop_count = groupedby_popular.count()["Price"]
pop_avg = groupedby_popular.mean()["Price"]
pop_sum = groupedby_popular.sum()["Price"]
popitemdf = pd.DataFrame({
    "Total Item Count": pop_count,
    "Item Price":pop_avg,
    "Tot Purchase Value":pop_sum
})
popitemdf = popitemdf.sort_values(by="Total Item Count", ascending=False)
popitemdf.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
popular_df = purchase_data.loc[:,["Item Name","Item ID", "Price"]]

groupedby_popular = popular_df.groupby(["Item ID","Item Name"])
groupedby_popular.head()
#do the counts
pop_count = groupedby_popular.count()["Price"]
pop_avg = groupedby_popular.mean()["Price"]
pop_sum = groupedby_popular.sum()["Price"]
popitemdf = pd.DataFrame({
    "Total Item Count": pop_count,
    "Item Price":pop_avg,
    "Tot Purchase Value":pop_sum
})
popitemdf = popitemdf.sort_values(by="Tot Purchase Value", ascending=False)
popitemdf.head()
```
